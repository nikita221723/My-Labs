#### (1)

Хотим проверить справедливость модели $$\log(S_n) = \log(S_0) + \widetilde{W}_{T_n}$$

где $n$ - номер временного интервала, 

 $S_n$ - цена акции в конце интервала $n$, 
 
 $T(n)$ -  объём сделок с начала наблюдений до конца интервала $n,$
 
   $\widetilde{W}_t = a t + \sigma W_t$ - броуновское движение со сдвигом. 

Задача: оценить параметры $\sigma$ и $a$ при помощи эмпирических методов поскольку:

$$\log(S_n) - \log(S_{n-1}) \overset{\text{d}}{=} a(T(n) - T(n-1)) + \sigma(T(n) - T(n-1))^{\frac{1}{2}}\xi, \ \xi \sim N(0, 1) $$ 

Тогда: $$\hat{a} = \frac{mean(\log(S_n) - \log(S_n - 1))}{mean(T(n) - T(n-1))}$$

$$\hat{\sigma^2} = var \left( \frac{\log(S_n) - \log(S_{n - 1}) - \hat{a}(T(n) - T(n-1))}{(T(n) - T(n-1))^{\frac{1}{2}}} \right)$$

   

```python
Sn = data['Close_csco'].to_numpy() #преобразуем в массивы 
T = data['Volume_csco'].to_numpy()
deltaSn = np.array([np.log(Sn[i]) - np.log(Sn[i-1]) for i in range(1, len(Sn))]) #создадим массивы приращений t_n - t_{n-1}
deltaT = T[1::]

a_hat = (np.mean(deltaSn)) / np.mean(deltaT) #считаем оценки 
sigma_hat = np.std((deltaSn - a_hat * deltaT) / np.sqrt(deltaT), ddof=1)
print(f'a_hat: {a_hat}')
print(f'sigma_hat: {sigma_hat}')
```

#### (2)

Промоделируйте процесс с оцененными параметрами и убедитесь, что траектории похожи

```python
import seaborn as sns
sns.relplot(Sn, #должны получить чет похожее
            height=6,
            aspect=2,
            color='black',
            kind='line').set(xlabel='timestamp', ylabel='real price', title='real stock price volatility')
```

```python
from scipy.stats import ks_2samp
LogSn_hat = [np.log(Sn[0])] + np.zeros_like(T) #инициализируем нашу логарифмированную траекторию 

np.random.seed(1111) #но 5829 дает наилучший пи-вэлью зуб даю 


for i in range(1, len(T)):
    LogSn_hat[i] = LogSn_hat[i-1] + a_hat * T[i-1] + sigma_hat * np.sqrt(T[i-1]) * np.random.normal(0, 1) #прибавляем приращение по известной нам формуле 

Snhat = np.exp(LogSn_hat) # экспонируем и получаем оценку траектории цен 

Trajectories = {'real prices': Sn, 'estimated prices': Snhat}

sns.relplot(Trajectories,
            height=6,
            aspect=2,
            kind='line').set(xlabel='timestamp', ylabel='price', title='stock price volatility')
```

Тест Смирнова 

```python
from scipy.stats import ks_2samp
diffs_sn_hat = np.diff(np.log(Snhat))
statistic, p_value = ks_2samp(diffs_sn_hat, deltaSn)
alpha = 0.05  # уровень значимости
if p_value < alpha:
    print("Отклоняем нулевую гипотезу: выборки имеют разные распределения.")
else:
    print("Не отклоняем нулевую гипотезу: выборки могут быть из одного распределения.")
p_value
```
